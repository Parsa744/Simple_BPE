```python
def find_most_friqunt_par(tlist):
  dict_of_pair = {}
  for i in range(len(tlist)-1):
    if tlist[i] + tlist[i+1] not in dict_of_pair:
      dict_of_pair[tlist[i] + tlist[i+1]] = {"freq": 1, "pairs": (tlist[i] , tlist[i+1])}
    else:
      dict_of_pair[tlist[i] + tlist[i+1]]["freq"] += 1
  try:
    most_freq_key = max(dict_of_pair, key=lambda k: dict_of_pair[k]["freq"])
  except:
    return None
  return dict_of_pair[most_freq_key]["pairs"]

def replace_pair(tlist,pair_list):
  i = 0
  while i < len(tlist)-1:
    print(i,len(tlist))
    if (tlist[i], tlist[i+1]) == pair_list:
      tlist[i] =pair_list[0]+pair_list[1]
      del tlist[i+1]
    else:
      i+=1

  return tlist


def BPE(text,merg_numb=5):
  list_text = list(text)
  print(list_text)
  most_friqunt_par = find_most_friqunt_par(list_text)
  list_text = replace_pair(list_text,most_friqunt_par)
  print(list_text)
  for i in range(merg_numb):
    last_len = len(list_text)
    most_friqunt_par = find_most_friqunt_par(list_text)
    list_text = replace_pair(list_text,most_friqunt_par)
    print(list_text)
    if len(list_text) ==  last_len:
      return list_text
  return list_text
```
